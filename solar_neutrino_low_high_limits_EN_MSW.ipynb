
# Solar neutrino survival probability: low- and high-energy limits

This notebook summarizes two important limiting cases for the electron-neutrino survival probability in solar neutrino physics:

1. **Low-energy limit:** vacuum oscillations averaged over many oscillation periods.
2. **High-energy limit:** matter-dominated MSW evolution inside the Sun.

We use the two-flavor approximation with the solar mixing parameters

$$
\theta \equiv \theta_{12},
\qquad
\Delta m^2 \equiv \Delta m^2_{21}.
$$

The main results are

$$
\boxed{
P_{ee}^{\rm low}
\simeq
1-\frac{1}{2}\sin^2 2\theta_{12}
}
$$

and

$$
\boxed{
P_{ee}^{\rm high}
\simeq
\sin^2\theta_{12}
}.
$$



## 0. Setup

For numerical estimates we use a typical illustrative value

$$
\sin^2\theta_{12} \simeq 0.307.
$$

If your fit uses a different value, change the variable `sin2_theta12` below.


```python

import numpy as np
import matplotlib.pyplot as plt

# Typical illustrative value for the solar mixing angle.
sin2_theta12 = 0.307

theta12 = np.arcsin(np.sqrt(sin2_theta12))
cos2_theta12_sq = 1.0 - sin2_theta12

print(f"sin^2(theta12) = {sin2_theta12:.4f}")
print(f"cos^2(theta12) = {cos2_theta12_sq:.4f}")
print(f"theta12 = {np.degrees(theta12):.2f} degrees")
```


# 1. Low-energy limit: vacuum-averaged oscillations

In vacuum, the two-flavor survival probability is

$$
P_{ee}
=
1
-
\sin^2 2\theta_{12}\,
\sin^2\left(
\frac{\Delta m^2_{21}L}{4E}
\right).
$$

Here:

- \(L\) is the propagation distance;
- \(E\) is the neutrino energy;
- \(\Delta m^2_{21}\) is the solar mass-squared splitting;
- \(\theta_{12}\) is the solar mixing angle.

The oscillation phase is

$$
\varphi
=
\frac{\Delta m^2_{21}L}{4E}.
$$

For solar neutrinos, this phase is often effectively averaged. This can happen because of the finite size of the production region in the Sun, the energy spread of the source, detector resolution, and the long Sun--Earth baseline.

The key average is

$$
\left\langle \sin^2\varphi \right\rangle
=
\frac{1}{2}.
$$

Therefore,

$$
\left\langle P_{ee}\right\rangle
=
1
-
\sin^2 2\theta_{12}\,
\left\langle \sin^2\varphi \right\rangle
=
1
-
\frac{1}{2}\sin^2 2\theta_{12}.
$$

Thus the low-energy, vacuum-averaged limit is

$$
\boxed{
P_{ee}^{\rm low}
\simeq
1-\frac{1}{2}\sin^2 2\theta_{12}
}.
$$

An equivalent form is

$$
P_{ee}^{\rm low}
=
\cos^4\theta_{12}
+
\sin^4\theta_{12}.
$$

This form is useful because it shows that after averaging, the interference term disappears. The survival probability is just the incoherent sum of the probabilities associated with the mass eigenstates.


```python

sin2_2theta12 = np.sin(2 * theta12)**2

Pee_low = 1.0 - 0.5 * sin2_2theta12
Pee_low_equivalent = np.cos(theta12)**4 + np.sin(theta12)**4

print("Low-energy / vacuum-averaged limit")
print("-----------------------------------")
print(f"sin^2(2 theta12) = {sin2_2theta12:.4f}")
print(f"Pee_low = 1 - 1/2 sin^2(2 theta12) = {Pee_low:.4f}")
print(f"Pee_low = cos^4(theta12) + sin^4(theta12) = {Pee_low_equivalent:.4f}")
```


## 1.1 Explicit check of the averaging

Since

$$
\sin^2\varphi
=
\frac{1-\cos 2\varphi}{2},
$$

the oscillating part averages to zero over many periods:

$$
\left\langle \cos 2\varphi \right\rangle = 0.
$$

Hence,

$$
\left\langle \sin^2\varphi \right\rangle
=
\frac{1}{2}.
$$

The short numerical check below shows this directly.


```python

phi = np.linspace(0, 100 * np.pi, 100_000)
average_sin2 = np.mean(np.sin(phi)**2)

print(f"<sin^2(phi)> = {average_sin2:.6f}")
```

```python

plt.figure(figsize=(8, 4))
plt.plot(phi[:2000], np.sin(phi[:2000])**2, label=r"$\sin^2\varphi$")
plt.axhline(0.5, linestyle="--", label=r"$\langle \sin^2\varphi\rangle = 1/2$")
plt.xlabel(r"Phase $\varphi$")
plt.ylabel(r"$\sin^2\varphi$")
plt.title("Averaging of the oscillating factor")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```


# 2. Matter effect in the Sun

In vacuum, the flavor states are related to the mass states by

$$
|\nu_e\rangle
=
\cos\theta_{12}|\nu_1\rangle
+
\sin\theta_{12}|\nu_2\rangle.
$$

Inside the Sun, electron neutrinos interact coherently with electrons in matter. This gives an additional matter potential

$$
V_e
=
\sqrt{2}\,G_F n_e,
$$

where \(G_F\) is the Fermi constant and \(n_e\) is the electron number density.

In the two-flavor approximation, the effective Hamiltonian in matter can be written, up to an irrelevant term proportional to the identity, as

$$
H_f
=
\frac{\Delta m^2_{21}}{4E}
\begin{pmatrix}
-\cos 2\theta_{12} & \sin 2\theta_{12} \\
\sin 2\theta_{12} & \cos 2\theta_{12}
\end{pmatrix}
+
\begin{pmatrix}
V_e & 0 \\
0 & 0
\end{pmatrix}.
$$

Because of the matter term, the mixing angle is modified:

$$
\theta_{12}
\longrightarrow
\theta_m.
$$

The effective mixing angle in matter satisfies

$$
\tan 2\theta_m
=
\frac{
\Delta m^2_{21}\sin 2\theta_{12}
}{
\Delta m^2_{21}\cos 2\theta_{12}
-
2EV_e
}.
$$

It is useful to define

$$
A
=
\frac{2EV_e}{\Delta m^2_{21}}.
$$

Then

$$
\tan 2\theta_m
=
\frac{\sin 2\theta_{12}}{\cos 2\theta_{12}-A}.
$$



# 3. What is the MSW resonance?

The MSW resonance is the condition under which the mixing angle in matter becomes maximal:

$$
\theta_m = \frac{\pi}{4}.
$$

Maximal mixing means

$$
\sin^2 2\theta_m = 1.
$$

Using

$$
\sin^2 2\theta_m
=
\frac{
\sin^2 2\theta_{12}
}{
(\cos 2\theta_{12}-A)^2+\sin^2 2\theta_{12}
},
$$

we see that the denominator is minimized when

$$
\boxed{
A = \cos 2\theta_{12}
}.
$$

Since

$$
A = \frac{2EV_e}{\Delta m^2_{21}},
$$

the resonance condition is

$$
\boxed{
2EV_e = \Delta m^2_{21}\cos 2\theta_{12}
}.
$$

Using

$$
V_e=\sqrt{2}G_F n_e,
$$

one can also write the resonance electron density as

$$
\boxed{
n_e^{\rm res}
=
\frac{
\Delta m^2_{21}\cos 2\theta_{12}
}{
2\sqrt{2}G_F E
}
}.
$$

## Physical meaning

At the resonance, the effective diagonal splitting in the flavor Hamiltonian is canceled by the matter potential. As a result, the two matter eigenstates become maximally mixed.

Important distinction:

- **MSW resonance** means \(A=\cos 2\theta_{12}\), so \(\theta_m=\pi/4\).
- **High-energy matter-dominated limit** means \(A\gg 1\), so \(\theta_m\to \pi/2\).

So the high-energy solar limit is not simply “being at resonance.” Rather, high-energy solar neutrinos are usually produced at densities above the resonance region and then evolve adiabatically through the changing solar density profile.


```python

# Mixing angle in matter as a function of A = 2 E V_e / Delta m^2

A = np.linspace(0, 5, 1000)

sin_2theta = np.sin(2 * theta12)
cos_2theta = np.cos(2 * theta12)

sin2_2theta_m = sin_2theta**2 / ((cos_2theta - A)**2 + sin_2theta**2)

A_res = cos_2theta

print("MSW resonance condition")
print("-----------------------")
print(f"A_res = cos(2 theta12) = {A_res:.4f}")
print("At this point, sin^2(2 theta_m) reaches 1.")
```

```python

plt.figure(figsize=(8, 5))
plt.plot(A, sin2_2theta_m, label=r"$\sin^2 2\theta_m$")
plt.axvline(A_res, linestyle="--", label=r"$A_{\rm res}=\cos 2\theta_{12}$")
plt.xlabel(r"$A = 2EV_e/\Delta m^2_{21}$")
plt.ylabel(r"$\sin^2 2\theta_m$")
plt.title("MSW resonance: maximal mixing in matter")
plt.ylim(0, 1.05)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```


# 4. High-energy solar limit: matter-dominated MSW evolution

For high-energy solar neutrinos, the matter term can dominate at the production point:

$$
A
=
\frac{2EV_e}{\Delta m^2_{21}}
\gg 1.
$$

In this limit, the effective mixing angle in matter tends to

$$
\theta_m \to \frac{\pi}{2}.
$$

This means that an electron neutrino produced deep in the Sun is approximately one matter eigenstate:

$$
|\nu_e\rangle
\approx
|\nu_2^m\rangle.
$$

If the solar density changes slowly enough, the propagation is **adiabatic**. Adiabatic evolution means that the neutrino stays on the same instantaneous matter-eigenstate branch while the density changes.

Therefore,

$$
|\nu_2^m\rangle
\longrightarrow
|\nu_2\rangle
$$

as the neutrino exits the Sun into vacuum.

Now the detector asks: what is the probability that the vacuum mass eigenstate \(|\nu_2\rangle\) is detected as an electron neutrino?

Since

$$
|\nu_e\rangle
=
\cos\theta_{12}|\nu_1\rangle
+
\sin\theta_{12}|\nu_2\rangle,
$$

we have

$$
\langle \nu_e|\nu_2\rangle
=
\sin\theta_{12}.
$$

Therefore,

$$
P_{ee}^{\rm high}
=
|\langle \nu_e|\nu_2\rangle|^2
=
\sin^2\theta_{12}.
$$

Thus the high-energy solar limit is

$$
\boxed{
P_{ee}^{\rm high}
\simeq
\sin^2\theta_{12}
}.
$$


```python

Pee_high = sin2_theta12

print("High-energy / MSW matter-dominated limit")
print("-----------------------------------------")
print(f"Pee_high = sin^2(theta12) = {Pee_high:.4f}")
```


## 4.1 Same result from the adiabatic formula

In the two-flavor adiabatic approximation,

$$
P_{ee}
=
\frac{1}{2}
+
\frac{1}{2}\cos 2\theta_{12}\cos 2\theta_m.
$$

In the high-energy matter-dominated limit,

$$
\theta_m\to\frac{\pi}{2},
$$

so

$$
\cos 2\theta_m \to \cos \pi = -1.
$$

Therefore,

$$
P_{ee}^{\rm high}
=
\frac{1}{2}
-
\frac{1}{2}\cos 2\theta_{12}.
$$

Using

$$
\sin^2\theta_{12}
=
\frac{1-\cos 2\theta_{12}}{2},
$$

we get

$$
\boxed{
P_{ee}^{\rm high}
=
\sin^2\theta_{12}
}.
$$


```python

Pee_high_adiabatic = 0.5 - 0.5 * np.cos(2 * theta12)

print("High-energy limit from the adiabatic formula")
print("---------------------------------------------")
print(f"cos(2 theta12) = {np.cos(2 * theta12):.4f}")
print(f"Pee_high = 1/2 - 1/2 cos(2 theta12) = {Pee_high_adiabatic:.4f}")
print(f"sin^2(theta12) = {sin2_theta12:.4f}")
```


# 5. Comparison of the two limits

The two useful limits are

$$
\boxed{
P_{ee}^{\rm low}
\simeq
1-\frac{1}{2}\sin^2 2\theta_{12}
}
$$

and

$$
\boxed{
P_{ee}^{\rm high}
\simeq
\sin^2\theta_{12}
}.
$$

The difference is physical:

- At **low energies**, matter effects are weak, and the vacuum oscillation phase is averaged.
- At **high energies**, matter effects dominate the production region, and adiabatic MSW evolution converts the produced \(\nu_e\) approximately into the vacuum mass eigenstate \(\nu_2\).

For \(\sin^2\theta_{12}\simeq 0.307\),

$$
P_{ee}^{\rm low}\approx 0.57,
\qquad
P_{ee}^{\rm high}\approx 0.31.
$$


```python

labels = ["low-energy\nvacuum average", "high-energy\nMSW limit"]
values = [Pee_low, Pee_high]

print("Numerical summary")
print("-----------------")
print(f"Pee_low  = {Pee_low:.4f}")
print(f"Pee_high = {Pee_high:.4f}")
print(f"Pee_low - Pee_high = {Pee_low - Pee_high:.4f}")

plt.figure(figsize=(7, 4))
plt.bar(labels, values)
plt.ylim(0, 1)
plt.ylabel(r"$P_{ee}$")
plt.title("Two limiting values of the solar electron-neutrino survival probability")
plt.grid(axis="y", alpha=0.3)

for i, value in enumerate(values):
    plt.text(i, value + 0.02, f"{value:.3f}", ha="center")

plt.show()
```


# 6. Optional: dependence on \(\sin^2\theta_{12}\)

Let

$$
x=\sin^2\theta_{12}.
$$

Then

$$
P_{ee}^{\rm high}=x.
$$

For the low-energy limit,

$$
\sin^2 2\theta_{12}
=
4\sin^2\theta_{12}\cos^2\theta_{12}
=
4x(1-x).
$$

Therefore,

$$
P_{ee}^{\rm low}
=
1-\frac{1}{2}\cdot 4x(1-x)
=
1-2x(1-x).
$$


```python

x = np.linspace(0, 1, 500)

P_low_x = 1 - 2 * x * (1 - x)
P_high_x = x

plt.figure(figsize=(8, 5))
plt.plot(x, P_low_x, label=r"$P_{ee}^{\rm low}=1-2x(1-x)$")
plt.plot(x, P_high_x, label=r"$P_{ee}^{\rm high}=x$")
plt.axvline(sin2_theta12, linestyle="--", label=fr"chosen $x={sin2_theta12:.3f}$")

plt.xlabel(r"$x=\sin^2\theta_{12}$")
plt.ylabel(r"$P_{ee}$")
plt.title(r"Low-energy and high-energy solar limits")
plt.ylim(0, 1)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```


# 7. Final conceptual summary

## Low-energy limit

The neutrino propagates approximately as in vacuum, but the oscillating term is averaged:

$$
\sin^2\left(
\frac{\Delta m^2_{21}L}{4E}
\right)
\rightarrow
\frac{1}{2}.
$$

Hence,

$$
P_{ee}^{\rm low}
\simeq
1-\frac{1}{2}\sin^2 2\theta_{12}.
$$

## MSW resonance

The MSW resonance occurs when the matter potential cancels the vacuum diagonal splitting:

$$
2EV_e
=
\Delta m^2_{21}\cos 2\theta_{12}.
$$

At this point,

$$
\theta_m=\frac{\pi}{4},
\qquad
\sin^2 2\theta_m=1.
$$

## High-energy limit

At high energies, the production point is matter-dominated:

$$
A=\frac{2EV_e}{\Delta m^2_{21}}\gg 1.
$$

Then

$$
\theta_m\to\frac{\pi}{2},
$$

and the produced electron neutrino approximately follows the matter eigenstate

$$
\nu_e\simeq \nu_2^m.
$$

Under adiabatic evolution,

$$
\nu_2^m\to \nu_2.
$$

Since the electron component of \(\nu_2\) is \(\sin\theta_{12}\), the survival probability becomes

$$
P_{ee}^{\rm high}
\simeq
|\langle \nu_e|\nu_2\rangle|^2
=
\sin^2\theta_{12}.
$$
