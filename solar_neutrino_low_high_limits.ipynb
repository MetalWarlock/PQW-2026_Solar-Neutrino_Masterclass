
# Solar neutrino survival probability: two useful limits

В этом ноутбуке оформлены два предельных случая для вероятности выживания электронного солнечного нейтрино:

1. **Низкие энергии / вакуумное усреднение**
\[
P_{ee}^{\rm low}\simeq 1-\frac12\sin^2 2\theta_{12}
\]

2. **Высокие энергии / MSW-предел в веществе Солнца**
\[
P_{ee}^{\rm high}\simeq \sin^2\theta_{12}
\]

Здесь \(\theta_{12}\) — солнечный угол смешивания.



## 0. Импорты и параметры

Для численных оценок возьмём типичное значение

\[
\sin^2\theta_{12}\approx 0.307.
\]

Если в твоём анализе используется другое значение, достаточно поменять переменную `sin2_theta12`.


```python

import numpy as np
import matplotlib.pyplot as plt

# Typical illustrative value. Replace by your fit value if needed.
sin2_theta12 = 0.307

theta12 = np.arcsin(np.sqrt(sin2_theta12))

print(f"sin^2(theta12) = {sin2_theta12:.4f}")
print(f"theta12 = {np.degrees(theta12):.2f} degrees")
```


# 1. Low-energy solar limit: vacuum-averaged oscillations

В вакууме для двух нейтринных состояний вероятность выживания электронного нейтрино имеет вид

\[
P_{ee}
=
1-\sin^2 2\theta_{12}\,
\sin^2\left(\frac{\Delta m^2_{21}L}{4E}\right).
\]

Здесь:

- \(L\) — расстояние распространения;
- \(E\) — энергия нейтрино;
- \(\Delta m^2_{21}\) — разность квадратов масс;
- \(\theta_{12}\) — угол смешивания.

Для солнечных нейтрино часто происходит усреднение по фазе осцилляций. Причины: конечный размер области рождения нейтрино в Солнце, распределение энергий, разрешение детектора и большой путь до Земли.

Фаза

\[
\varphi = \frac{\Delta m^2_{21}L}{4E}
\]

быстро меняется, поэтому

\[
\left\langle \sin^2\varphi \right\rangle = \frac12.
\]

Тогда

\[
\left\langle P_{ee}\right\rangle
=
1-\sin^2 2\theta_{12}\cdot \frac12.
\]

И получаем низкоэнергетический предел:

\[
\boxed{
P_{ee}^{\rm low}
\simeq
1-\frac12\sin^2 2\theta_{12}
}
\]

Эквивалентная форма:

\[
1-\frac12\sin^2 2\theta
=
\cos^4\theta+\sin^4\theta.
\]

Эта форма показывает физический смысл: после усреднения интерференционный член пропадает, и остаются только вероятности быть в массовых состояниях \(\nu_1\) и \(\nu_2\).


```python

# Low-energy vacuum-averaged survival probability

sin2_2theta12 = np.sin(2 * theta12)**2
Pee_low = 1 - 0.5 * sin2_2theta12

# Equivalent expression
cos2_theta12 = np.cos(theta12)**2
Pee_low_equiv = cos2_theta12**2 + sin2_theta12**2

print("Low-energy / vacuum-averaged limit")
print("-----------------------------------")
print(f"sin^2(2 theta12) = {sin2_2theta12:.4f}")
print(f"Pee_low = 1 - 1/2 sin^2(2 theta12) = {Pee_low:.4f}")
print(f"Pee_low = cos^4(theta12) + sin^4(theta12) = {Pee_low_equiv:.4f}")
```


## Проверка усреднения \(\langle \sin^2\varphi\rangle = 1/2\)

Возьмём много значений фазы \(\varphi\) на большом числе периодов и явно усредним \(\sin^2\varphi\).


```python

phi = np.linspace(0, 100 * np.pi, 100_000)
average_sin2 = np.mean(np.sin(phi)**2)

print(f"<sin^2(phi)> = {average_sin2:.6f}")
```

```python

plt.figure(figsize=(8, 4))
plt.plot(phi[:2000], np.sin(phi[:2000])**2, label=r"$\sin^2\varphi$")
plt.axhline(0.5, linestyle="--", label=r"average $=1/2$")
plt.xlabel(r"Phase $\varphi$")
plt.ylabel(r"$\sin^2\varphi$")
plt.title(r"Averaging of the oscillating factor")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```


# 2. High-energy solar limit: MSW matter-dominated regime

Для высокоэнергетических солнечных нейтрино главным становится не вакуумная формула, а взаимодействие нейтрино с веществом Солнца.

Электронное нейтрино \(\nu_e\) имеет дополнительное когерентное взаимодействие с электронами среды. Это даёт потенциал вещества

\[
V_e = \sqrt{2}G_F n_e,
\]

где \(n_e\) — электронная плотность.

Из-за этого угол смешивания в веществе становится другим:

\[
\theta_{12} \rightarrow \theta_m.
\]

В двухфлейворном приближении

\[
\tan 2\theta_m
=
\frac{\Delta m^2_{21}\sin 2\theta_{12}}
{\Delta m^2_{21}\cos 2\theta_{12} - 2EV_e}.
\]

Удобно ввести безразмерный параметр

\[
A =
\frac{2EV_e}{\Delta m^2_{21}}.
\]

Тогда

\[
\tan 2\theta_m
=
\frac{\sin 2\theta_{12}}
{\cos 2\theta_{12}-A}.
\]

При высоких энергиях \(E\) параметр \(A\) большой:

\[
A\gg 1.
\]

В этом пределе

\[
\theta_m \rightarrow \frac{\pi}{2}.
\]

То есть электронное нейтрино, рождённое в плотной области Солнца, почти совпадает с собственным состоянием вещества:

\[
|\nu_e\rangle \approx |\nu_2^m\rangle.
\]

Если эволюция адиабатическая, нейтрино не перескакивает между ветвями собственных состояний. Тогда при выходе из Солнца

\[
|\nu_2^m\rangle \longrightarrow |\nu_2\rangle.
\]

Детектор на Земле измеряет вероятность найти электронное нейтрино в состоянии \(|\nu_2\rangle\):

\[
P_{ee}^{\rm high}
=
|\langle \nu_e|\nu_2\rangle|^2.
\]

А в вакууме

\[
|\nu_e\rangle
=
\cos\theta_{12}|\nu_1\rangle
+
\sin\theta_{12}|\nu_2\rangle.
\]

Следовательно,

\[
\langle \nu_e|\nu_2\rangle = \sin\theta_{12},
\]

и поэтому

\[
\boxed{
P_{ee}^{\rm high}
\simeq
\sin^2\theta_{12}
}
\]


```python

# High-energy MSW limit

Pee_high = sin2_theta12

print("High-energy / MSW limit")
print("-----------------------")
print(f"Pee_high = sin^2(theta12) = {Pee_high:.4f}")
```


## То же самое через адиабатическую формулу

В адиабатическом приближении для двух состояний можно записать

\[
P_{ee}
=
\frac12
+
\frac12\cos 2\theta_{12}\cos 2\theta_m.
\]

В высокоэнергетическом пределе

\[
\theta_m\rightarrow \frac{\pi}{2},
\]

поэтому

\[
\cos 2\theta_m \rightarrow \cos \pi = -1.
\]

Тогда

\[
P_{ee}^{\rm high}
=
\frac12
-
\frac12\cos 2\theta_{12}.
\]

Используем тождество

\[
\sin^2\theta = \frac{1-\cos 2\theta}{2}.
\]

Получаем

\[
\boxed{
P_{ee}^{\rm high}
=
\sin^2\theta_{12}
}
\]


```python

# High-energy limit using the adiabatic formula

cos_2theta12 = np.cos(2 * theta12)
cos_2theta_m_high = -1.0

Pee_high_adiabatic = 0.5 + 0.5 * cos_2theta12 * cos_2theta_m_high

print("High-energy limit from adiabatic formula")
print("----------------------------------------")
print(f"cos(2 theta12) = {cos_2theta12:.4f}")
print(f"cos(2 theta_m) -> {cos_2theta_m_high:.1f}")
print(f"Pee_high = 1/2 + 1/2 cos(2 theta12) cos(2 theta_m) = {Pee_high_adiabatic:.4f}")
print(f"sin^2(theta12) = {sin2_theta12:.4f}")
```


# 3. Comparison of the two limits

Итого:

\[
\boxed{
P_{ee}^{\rm low}
\simeq
1-\frac12\sin^2 2\theta_{12}
}
\]

\[
\boxed{
P_{ee}^{\rm high}
\simeq
\sin^2\theta_{12}
}
\]

Смысл различия:

- **низкие энергии:** вещество почти не влияет, осцилляции в вакууме усредняются;
- **высокие энергии:** вещество Солнца доминирует, \(\nu_e\) адиабатически превращается в почти чистое \(\nu_2\).

Поэтому низкоэнергетический предел обычно около \(0.55-0.60\), а высокоэнергетический — около \(0.30\).


```python

print("Summary")
print("-------")
print(f"Pee_low  = {Pee_low:.4f}")
print(f"Pee_high = {Pee_high:.4f}")
print(f"Difference Pee_low - Pee_high = {Pee_low - Pee_high:.4f}")
```

```python

labels = ["low-energy\nvacuum avg.", "high-energy\nMSW"]
values = [Pee_low, Pee_high]

plt.figure(figsize=(6, 4))
plt.bar(labels, values)
plt.ylim(0, 1)
plt.ylabel(r"$P_{ee}$")
plt.title(r"Two solar-neutrino survival probability limits")
plt.grid(axis="y", alpha=0.3)

for i, value in enumerate(values):
    plt.text(i, value + 0.02, f"{value:.3f}", ha="center")

plt.show()
```


# 4. Optional: dependence on \(\sin^2\theta_{12}\)

Ниже построим оба предела как функции от \(\sin^2\theta_{12}\).

Пусть

\[
x = \sin^2\theta_{12}.
\]

Тогда

\[
P_{ee}^{\rm high}=x.
\]

А для низких энергий:

\[
\sin^2 2\theta_{12}
=
4\sin^2\theta_{12}\cos^2\theta_{12}
=
4x(1-x).
\]

Поэтому

\[
P_{ee}^{\rm low}
=
1-\frac12\cdot 4x(1-x)
=
1-2x(1-x).
\]


```python

x = np.linspace(0, 1, 500)

P_low_x = 1 - 2 * x * (1 - x)
P_high_x = x

plt.figure(figsize=(8, 5))
plt.plot(x, P_low_x, label=r"$P_{ee}^{low}=1-2x(1-x)$")
plt.plot(x, P_high_x, label=r"$P_{ee}^{high}=x$")
plt.axvline(sin2_theta12, linestyle="--", label=fr"chosen $x={sin2_theta12:.3f}$")

plt.xlabel(r"$x=\sin^2\theta_{12}$")
plt.ylabel(r"$P_{ee}$")
plt.title(r"Low-energy and high-energy solar limits")
plt.ylim(0, 1)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```


# 5. Final answer in words

**Low energy:**  
Солнечные нейтрино ведут себя почти как в вакууме, но осцилляционная фаза много раз пробегает период. Поэтому

\[
\sin^2\left(\frac{\Delta m^2_{21}L}{4E}\right)
\rightarrow
\frac12.
\]

Отсюда

\[
P_{ee}^{\rm low}
\simeq
1-\frac12\sin^2 2\theta_{12}.
\]

**High energy:**  
Внутри Солнца вещество сильно меняет смешивание. При больших \(E\)

\[
\theta_m\rightarrow \frac{\pi}{2},
\]

поэтому рождённое \(\nu_e\) почти совпадает с \(\nu_2^m\). При адиабатическом выходе из Солнца

\[
\nu_2^m\rightarrow \nu_2.
\]

А вероятность зарегистрировать \(\nu_2\) как электронное нейтрино равна

\[
|\langle \nu_e|\nu_2\rangle|^2
=
\sin^2\theta_{12}.
\]

Поэтому

\[
P_{ee}^{\rm high}
\simeq
\sin^2\theta_{12}.
\]
